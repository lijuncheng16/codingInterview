# 1557. Minimum Number of Vertices to Reach All Nodes


Given a directed acyclic graph, with n vertices numbered from 0 to n-1, and an array edges where edges[i] = [fromi, toi] represents a directed edge from node fromi to node toi.

Find the smallest set of vertices from which all nodes in the graph are reachable. It's guaranteed that a unique solution exists.

Notice that you can return the vertices in any order.
GPT3.5 and GPT4 remembered the counting in-degree solution, but failed to actually do it with dfs or others, probably haven't seen it in training data.


```python
class Solution:
    def dfs(self, vertex, first=False):
        if vertex in self.reachable:
            return
        if not first:
            self.reachable.add(vertex)
            first = False
        for _vertex in self.adj_matrix[vertex]:
            self.dfs(_vertex)

    def findSmallestSetOfVertices(self, n: int, edges: List[List[int]]) -> List[int]:
        self.adj_matrix = defaultdict(list)
        self.reachable = set()
        for edge in edges:
            self.adj_matrix[edge[0]].append(edge[1])
        
        all_nodes = set()
        for i in range(n):
            all_nodes.add(i)
            self.dfs(i, True)
        
        return list(all_nodes - self.reachable)
```

```python
def findSmallestSetOfVertices(self, n: int, edges: List[List[int]]) -> List[int]:
    graph = defaultdict(set)
    visited = set()
    res = set()

    for i,j in edges:
        graph[i].add(j)

    def dfs(node):
        visited.add(node)
        for child in graph[node]:
            if child not in visited:
                dfs(child)
            elif child in res:
                res.remove(child)

    for i in range(n):
        if i not in visited:
            res.add(i)
            dfs(i)
    return res
```

```python
def findSmallestSetOfVertices(self, n: int, edges: List[List[int]]) -> List[int]:
    all_nodes = set(range(n))

    # For each edge, mark the destination node (node with incoming edge).
    for _, to in edges:
        if to in all_nodes:
            all_nodes.remove(to)

    return list(all_nodes)

```
