# Bag of Tokens 

```python
from typing import List
```

```python

def bagOfTokensScore(tokens: List[int], P: int) -> int:
    tokens = sorted(tokens)
    print(tokens)
    length = len(tokens)
    l = 0
    r = length -1 
    score = 0
    max_score = 0 
    if tokens == []:
        return 0
    else:
        if length == 1:
            if P> tokens[0]:
                return 1
        if P < tokens[0]:
            return 0
        while l< r: 
            while P >= tokens[l]:
                P = P -tokens[l]
                score += 1
                max_score = max(score, max_score)
                l +=1
                print(f'P:{P}, score:{score}, l:{l}')
            if score >= 1:
                P = P + tokens[r]
                score -= 1
                r -= 1
                print(f'P:{P}, score:{score}, r:{r}')
            if r-l < 1:
                break
    print(max_score) 
```

```python
tokens = [26]
P = 50
bagOfTokensScore(tokens,P)
```

```python
tokens = []
P = 50
bagOfTokensScore(tokens,P)
```

h index

```python
def hIndex(citations: List[int]) -> int:
    citations = sorted(citations)
    print(citations)
    n_paper = len(citations)
    h = 0
    if citations == []:
        return 0
    if citations[-1]==0:
        return h
    while citations [-1-h] >= h:
        print(f'h_index:{h}, {citations[-1-h]}')
        h += 1
        if h >= n_paper:
            break
        if citations[-1-h] <= h:
            break
    return h
```

```python
citations = [3,0,6,1,5]
hIndex(citations)
```

```python
def isHappy(n: int) -> bool:
        digits = str(n)
        new_number = 0
        for digit in digits:
            new_number += int(digit) **2
        print(new_number, new_number == n**2, n)
        if new_number == 1:
            return True
        if new_number == n:
            print('false')
            return False
        return(isHappy(new_number))
```

```python
print(isHappy(2))
```
